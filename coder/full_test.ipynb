# 第一轮pass1

```python
from evaluate.pass1_direct import Pass1
import json
from collections import OrderedDict

dataset_name = 'humaneval'

def load_data(file_path):
    dataset = {}
    code = {}
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            record = json.loads(line)
            code[record['id']] = record['code']
    with open(f'../benchmark/ourbench/{dataset_name}_dyn.jsonl', 'r', encoding='utf-8') as file:
        for line in file:
            record = json.loads(line)
            if record['task_id'].split('/')[1].split('-')[1]=='2':
                task_id = record['task_id'].split('/')[1].split('-')[0]
                task = record['raw_problem']
                new_code = code[task_id]
                test_list = record['raw_test_input']
                dataset[task_id] = [task, new_code, test_list]
    return dataset


error = []
method_list = [
    'agent_deepseek-v3',
    'agent_gpt-4o-mini',
    'gpt_direct',
    'gpt_COT',
    'gpt_selfplanning',
    'qwen',
    'ds',
    'claude'
]

import logging
logging.basicConfig(filename=f'test/{dataset_name}/logs/pass_latest.log', level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

for method in method_list:
    id_list = list(load_data(f'test/{dataset_name}/code/ds_1.jsonl').keys())
    pass_list = {}
    for id in id_list:
        pass_list[id] = 5
    for i in range(1,6):
        dataset_agent = load_data(f'test/{dataset_name}/code/{method}_{i}.jsonl')
        class eval_agent(Pass1):
            def __init__(self):
                super().__init__()
                self.slient = True

            def load_data(self):
                return dataset_agent
        eval = eval_agent()
        error_list, failed_list, error_type_list = eval.run()
        error.append(error_list)
        for id in failed_list:
            pass_list[id] -= 1

        with open(f'test/{dataset_name}/error_detail/{method}_{i}_error_raw.jsonl', 'w', encoding='utf-8') as file:
            for id in sorted(error_type_list.keys()):
                record = error_type_list[f'{id}']
                file.write(json.dumps({'id': f'{id}', 'result': record['result'], 'error_type': record['error_type']})+'\n')
    print(pass_list)
    pass_1 = sum(pass_list.values()) / len(pass_list) /5
    print(pass_1)
    # 计算pass_list中值不为0的比例
    pass_5 = sum([1 for i in pass_list.values() if i !=0]) / len(pass_list)
    print(pass_5)
    pass_3 = 0
    for num in pass_list.values():
        if num >= 3:
            pass_3 += 1
        if num == 2:
            pass_3 += 0.9
        if num == 1:
            pass_3 += 0.6

    pass_3 = pass_3 / len(pass_list)
    print(pass_3)

    logger.info(f"--------{method}-v1-latest--------")
    logger.info(f"pass_1: {pass_1}")
    logger.info(f"pass_3: {pass_3}")
    logger.info(f"pass_5: {pass_5}")
    logger.info(f"pass_list: {pass_list}")
```
